```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat
from scipy.ndimage.filters import gaussian_filter

# Define paths
image_path = "/Users/akshara/Documents/info/ShanghaiTech/part_A/train_data/images"
density_path = "/Users/akshara/Documents/info/ShanghaiTech/part_A/train_data/ground-truth"

# Load image and corresponding GT
sample_image = os.listdir(image_path)[0]
gt_file = 'GT_' + sample_image.replace('.jpg', '.mat')

# Load annotation data
gt_data = loadmat(os.path.join(density_path, gt_file))
points = gt_data["image_info"][0, 0][0, 0][0]  # list of (x, y) coordinates

# Read image
img = cv2.imread(os.path.join(image_path, sample_image))
h, w = img.shape[:2]

# Create empty density map
density_map = np.zeros((h, w), dtype=np.float32)

# Mark each point on the density map
for point in points:
    x, y = min(int(point[0]), w - 1), min(int(point[1]), h - 1)
    density_map[y, x] = 1

# Apply Gaussian filter for smooth density visualization
density_map = gaussian_filter(density_map, sigma=15)

# Display
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(density_map, cmap='jet')
plt.title("Generated Density Map")
plt.axis("off")

plt.show()

print(f"Estimated People Count: {np.sum(density_map):.0f}")
```

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat
from scipy.ndimage import gaussian_filter
import random

# Paths
image_path = "/Users/akshara/Documents/info/ShanghaiTech/part_A/train_data/images"
density_path = "/Users/akshara/Documents/info/ShanghaiTech/part_A/train_data/ground-truth"

# -------------------------
# CHOOSE ANY IMAGE HERE
# -------------------------

sample_image = "IMG_14.jpg"
print("Using:", sample_image)

# Load ORIGINAL BGR image
img_bgr = cv2.imread(os.path.join(image_path, sample_image))

# Convert to RGB
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

# Resize to 256×256
resized = cv2.resize(img_rgb, (256, 256))

# Downscale to half size
downscaled = cv2.resize(img_rgb, 
                        (img_rgb.shape[1]//2, img_rgb.shape[0]//2))

# ---------------------------------------
# Load Ground Truth for Density Map
# ---------------------------------------
gt_file = "GT_" + sample_image.replace(".jpg", ".mat")
gt = loadmat(os.path.join(density_path, gt_file))
points = gt["image_info"][0,0][0,0][0]

h, w = img_rgb.shape[:2]
density_map = np.zeros((h, w), dtype=np.float32)

for p in points:
    x = min(int(p[0]), w-1)
    y = min(int(p[1]), h-1)
    density_map[y, x] = 1

density_map = gaussian_filter(density_map, sigma=10)

# ---------------------------------------
# Display ALL IMAGES
# ---------------------------------------
plt.figure(figsize=(15, 10))

plt.subplot(2, 3, 1)
plt.imshow(cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB))
plt.title("Original BGR Image")
plt.axis("off")

plt.subplot(2, 3, 2)
plt.imshow(img_rgb)
plt.title("RGB Image")
plt.axis("off")

plt.subplot(2, 3, 3)
plt.imshow(resized)
plt.title("Resized (256×256)")
plt.axis("off")

plt.subplot(2, 3, 4)
plt.imshow(downscaled)
plt.title("Downscaled (Half Size)")
plt.axis("off")

plt.subplot(2, 3, 5)
plt.imshow(img_rgb)
plt.title("Original RGB (Reference)")
plt.axis("off")

plt.subplot(2, 3, 6)
plt.imshow(density_map, cmap="jet")
plt.title("Generated Density Map")
plt.axis("off")

plt.show()

# ---------------------------------------
# Head Count
# ---------------------------------------
head_count = np.sum(density_map)
print("Estimated Head Count:", int(head_count))
```
